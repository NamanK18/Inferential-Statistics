# Inferential Statistics:-
#### It is used together ,some information about the population from the sample of data collected.There are many tests which are used in inferential statistics,depending on the objective and the data.
#### Objective 1:-To find out ,is there a significant difference in the marks of students between section A and section B.For any objective we formulate a hypothesis.
#### Null Hypothesis(Ho)=There is no significant difference in the marks between students of section A and B.
#### Alternate Hypothesis(Ha)=There is a significant difference in the marks of Section A and B.
### The test to be used:-
#### Step1- Check the normality for the variable wingA and WingB separately.
#### Step2- If both the variable have normal distribution then we use a parametric test known as 'Independent sample t-Test'.If the variables don't have the normal distribution then we use a non-paramtric test known as 'Mann Whitney U Test'.
### How to do Independent Sample t-Test in python:-
#### Import the library and import the dataset into python.
#### To do the Independent Sample t-Test in python,the function used is ttest_ind.
### How to import a specific function from the library:-
#### TO import the function 'ttest_ind' from the python library 'scipy.stats'

```python
import pandas as pd
```

```python
df=pd.read_csv('D:\Section.csv')
```

#### There are two variables to be used for dataframe.The independent sample t_test will give a t value and this t value has to be compared with the t critical value.

#### t_statistics,p_value=stats.ttest_ind(df['wingA'],df['wingB']) #to calculate t_statistics
##### type t_statistics and run the code.

#### The value from the python is the t calculated.This value has to be compared wih the critical value of t.This critical value will be obtained from the t table. Use the column for 5% significance level.Degreee of freedom =(n-2) where n is the sample size.
#### Get the Crtitical value from the statistical table.

## 1.) If the calculated t from python is less than the critical value from the table then you accept the Null Hypothesis and conclude there is no difference in the performance of the students between the two wings.
## 2.) If the calculated t from python is greater than the critical value from the table then you accept the Alternate Hypothesis and conclude there is a difference in the performance of the students between the two wings.
